```python
df['prix_fonctionnalite_av_O'].value_counts(dropna=False)

def map_prix(x):
    # 1) nettoyer / convertir en float
    if pd.isna(x):
        return -1
    s = str(x).lower()
    if 'gratuit' in s or s in ['0', '0.0', '0€']:
        return 0
    # extraire le nombre
    import re
    m = re.search(r'(\d+\.?\d*)', s)
    if not m:
        return -1
    v = float(m.group(1))

    # 2) catégoriser
    if v < 2:
        return 0
    elif 2 <= v <= 7:
        return 1
    elif 8 <= v <= 34:
        return 2
    elif v >= 35:
        return 3
    else:
        return -1

df['prix_fonctionnalite_av_O'] = df['prix_fonctionnalite_av_O'].apply(map_prix)
df['prix_fonctionnalite_av_O'].value_counts(dropna=False)
```

```python

df['prix_fonctionnalite_av_N'] = df['prix_fonctionnalite_av_N'].apply(map_prix)
df['prix_fonctionnalite_av_N'].value_counts(dropna=False)
```

Mapping pour les tranches d'ages des personnes connaissant le brainstorming ( age_O) et ceux qui ne connaisent pas ( age _N)

```python
#Pour l'age_O

mapping_mid = {
    '18-25':      21.5,
    '26-30':      28.0,
    '31-35':      33.0,
    '36-40':      38.0,
    '41-45':      43.0,
    '46-50':      48.0,
    '51-55':      53.0,
    '56 et plus': 60.0,   # ou 56, ou la moyenne d’une estimation [56–65]
}
df['age_O'] = (
    df['age_O']
      .map(mapping_mid)    # remplace les tranches par le milieu
      .fillna(-1)          # NaN devient -1
)

df['age_O'].value_counts(dropna=False)

```

```python
#Meme chose pour l'age_N
df['age_N'] = (
    df['age_N']
      .map(mapping_mid)    # remplace les tranches par le milieu
      .fillna(-1)          # NaN devient -1
)

df['age_N'].value_counts(dropna=False)
```
