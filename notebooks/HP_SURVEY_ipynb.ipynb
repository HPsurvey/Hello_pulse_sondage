# APPLICATION DU MAPPING ( histoire de pouvoir uniformiser les réponses )

```python
# Inspection des valeurs réelles des habitudes de travail
df['hbt_travail'].value_counts(dropna=False)
```

```python
mapping_hbt_travail = {
    'En groupe':0,
    'Individuellement':1,
    'Individuellement;En groupe' : 2
}
df['hbt_travail'] = df['hbt_travail'].map(mapping_hbt_travail)
```

```python
# Inspection des valeurs réelles des habitudes de travail
df['hbt_travail'].value_counts(dropna=False)
```

```python
#Inspection des valeurs réelles de l'utilisation du brainstorming
df['uti_appli_brainstorming'].value_counts(dropna=False)
```

```python
mapping_uti_appli_brainstorming = {
    'Oui':1,
    'Non':0,
    'Non mais j’utilise mes outils de brainstorming': 0,
    'Oui et j’utilise aussi les outils de brainstorming':1
}
df['uti_appli_brainstorming'] = df['uti_appli_brainstorming'].map(mapping_uti_appli_brainstorming)
```

```python
#Inspection des valeurs réelles de l'utilisation du brainstorming
df['uti_appli_brainstorming'].value_counts(dropna=False)
```
